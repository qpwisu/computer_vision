# 데이터 읽기
- keras로부터 CIFAR 10 데이터을 읽기

```
from keras.datasets import cifar10
(cx_train, cy_ftrain), (cx_test, cy_test) = cifar10.load_data()
```

## cx_train[40] 영상을 img 변수로 저장하고 plt.imshow 함수를 이용하여 가시화기 위한 코드를 작성하시오

```
(cx_train, cy_train), (cx_test,cy_test) = cifar10.load_data()
img = cx_train[40]
plt.imshow(img)

def histogram_256(img):
  R_channel = img[:, :, 0]
  G_channel = img[:, :, 1]
  B_channel = img[:, :, 2]
  Rhist_256, Rbins_256 = np.histogram(
      R_channel,bins=256,range=(np.min(R_channel),np.max(R_channel)))
  Ghist_256, Gbins_256 = np.histogram(
      G_channel,bins=256,range=(np.min(G_channel),np.max(G_channel)))
  Bhist_256, Bbins_256 = np.histogram(
      B_channel,bins=256,range=(np.min(B_channel),np.max(B_channel)))
  Rhist_norm_256, Rbins_norm_256 = np.histogram(
      R_channel,bins=256,range=(np.min(R_channel),np.max(R_channel)),normed=True)
  Ghist_norm_256, Gbins_norm_256 = np.histogram(
      G_channel,bins=256,range=(np.min(G_channel),np.max(G_channel)),normed=True)
  Bhist_norm_256, Bbins_norm_256 = np.histogram(
      B_channel,bins=256,range=(np.min(B_channel),np.max(B_channel)),normed=True)
def normal_histogram_256(img):
  R_channel = img[:, :, 0]
  G_channel = img[:, :, 1]
  B_channel = img[:, :, 2]
  Rhist_norm_256, Rbins_norm_256 = np.histogram(
      R_channel,bins=256,range=(np.min(R_channel),np.max(R_channel)),normed=True)
  Ghist_norm_256, Gbins_norm_256 = np.histogram(
      G_channel,bins=256,range=(np.min(G_channel),np.max(G_channel)),normed=True)
  Bhist_norm_256, Bbins_norm_256 = np.histogram(
      B_channel,bins=256,range=(np.min(B_channel),np.max(B_channel)),normed=True)

histogram_256(img)
normal_histogram_256(img)

plt.figure(figsize=(16,8))

#Subplot 1
plt.subplot(2,3,1)
plt.bar(Rbins_256[:-1],Rhist_256,width=1)
plt.title('RHistogram_256')
plt.subplot(2,3,2)
plt.bar(Gbins_256[:-1],Ghist_256,width=1)
plt.title('GHistogram_256')
plt.subplot(2,3,3)
plt.bar(Bbins_256[:-1],Bhist_256,width=1)
plt.title('BHistogram_256')

plt.subplot(2,3,4)
plt.bar(Rbins_norm_256[:-1],Rhist_norm_256,width=1)
plt.title('Normal RHistogram_256')
plt.subplot(2,3,5)
plt.bar(Gbins_norm_256[:-1],Ghist_norm_256,width=1)
plt.title('Normal GHistogram_256')
plt.subplot(2,3,6)
plt.bar(Bbins_norm_256[:-1],Bhist_norm_256,width=1)
plt.title('Normal BHistogram_256')
plt.show()
```
